## Section 2 — Import Libraries

```python
import pandas as pd
import numpy as np
```

## Section 3 — Load Raw Data

```python
stores= pd.read_csv(r'C:\Users\Administrator\.ipython\stores_dirty.csv')
products= pd.read_csv(r'C:\Users\Administrator\.ipython\products_dirty.csv')
customers= pd.read_csv(r'C:\Users\Administrator\.ipython\customers_dirty.csv')
sales= pd.read_csv(r'C:\Users\Administrator\.ipython\sales_dirty.csv')
inventory= pd.read_csv(r'C:\Users\Administrator\.ipython\inventory_dirty.csv')
```

## Section 4 — Inspect Data

```python
print(stores.head())
print(products.head())
print(customers.head())
print(sales.head())
print(inventory.head())
```

## Section 5 — Check Missing Values

```python
print(stores.isna().sum())
print(products.isna().sum())
print(customers.isna().sum())
print(sales.isna().sum())
print(inventory.isna().sum())
```

## Section 6 — Remove Duplicate Rows

```python
sales = sales.drop_duplicates()
```

```python
print(sales.duplicated().sum())
```

## Section 7 — Fix Inconsistent Text

```python
stores['City'].unique()
```

```python
stores['City']= stores['City'].str.title()
```

## Section 8 — Fix Currency Values

```python
products['Selling_Price'] = products['Selling_Price'].astype(str).str.replace('$', '', regex= False)

products['Selling_Price'] = products['Selling_Price'].astype(float)
```

## Section 9 — Fix Negative Values

```python
customers= customers[customers['Age'] > 0]
sales= sales[sales['Quantity'] > 0]
inventory= inventory[inventory['Stock_Quantity']>= 0]
```

## Section 10 — Handle Missing Values

```python
customers= customers['Email'].fillna('unknown@email.com')
sales= sales['Discount'].fillna(0)
```

## Section 11 — Fix Data Types

```python
print(sales.columns)
print(customers.columns)
```

```python
sales.columns = sales.columns.str.strip()
customers.columns = customers.columns.str.strip()
```

```python
sales['Date']= pd.to_datetime(sales['Date'])
customers['Signup_Date']= pd.to_datetime(customers['Signup_Date'])
```

## Section 12 — Create Clean Revenue Column

```python
sales['Revenue']= sales['Quantity'] * sales['Unit_Price']
```

## Section 13 — Save Clean Data

```python
inventory.to_csv(r'cleaned\inventory_clean.csv', index=False)
```
